```python
#function to get first 3 names from cast
import ast
def get_cast(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            L.append(i['name'])
            counter=counter+1
        else:
            break
    return L
```

```python
movies['cast']= movies['cast'].apply(get_cast)
```

```python
import ast
def get_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i["job"]== 'Director':
            L.append(i['name'])
    return L
```

```python
movies['crew']= movies['crew'].apply(get_director)
```

### Text Vectorization -  We will use Text Vectorization in which we will convert all Text(Tags) into Vector
#### There are many techniques for Text Vectorization - 1. Bags of words 2. IFIDF etc
#### we will be using Bags of words - 
#### exception We will be Neglecting some STOPWORDS from STRING first - e,g to, too, be, is, are, the etc
###### we will use 5000 words of bag in this algoritham (we can use any count of word more the count more the algoritham takes time )

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors[0]
```

```python
import numpy as np
np.set_printoptions(threshold=np.inf)
```

```python
#Complte upto 5000 words "bags of words"
cv.get_feature_names_out()
```

```python
# but some name are reapeating like eg ape apes, agent, agents, age , aged, allow allows, loving, loved, loves
# we have to neglect them as well because they almost have similar meanings

import nltk # library
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

ps.stem('loved')
```

```python
ps.stem('allows')
```

```python
ps.stem('allowed')
```

```python
# now we will apply it for every row tags with function
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags']= new_df['tags'].apply(stem)
```

```python
# Now we will seperate movies with angle of vector between 2 movies
# basically we are creating angles of one movie with all other [4900] movies
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity= cosine_similarity(vectors)
```

```python
similarity[2]
```

```python
similarity[0]
```

```python
# Find Index positions of movie

new_df[new_df['title']=='Batman Begins'].index[0]
```

```python
#we are sorting with Index positon and getting similar movie results upto 6 movies
sorted(list(enumerate(similarity[0])), reverse=True, key= lambda x:x[1])[1:6]
```

```python
def recommend(movie):
        movie_index= new_df[new_df['title']==movie].index[0]
        distances = similarity[movie_index]
        movies_list = sorted(list(enumerate(distances)), reverse=True, key= lambda x:x[1])[1:6]

        for i in movies_list:
            print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```

```python
recommend('Avatar')
```

```python
recommend('The Dark Knight Rises')
```
